```python
import pandas as pd
from kafka import KafkaProducer
import json
import time

# Load dataset
df = pd.read_csv(r"C:\Water_Monitoring_System\data\brisbane_water_quality.csv")

# Kafka Producer
producer = KafkaProducer(
    bootstrap_servers='localhost:9092',
    value_serializer=lambda v: json.dumps(v).encode('utf-8')
)


batch_size = 100   

for i in range(0, len(df), batch_size):
    batch = df.iloc[i:i+batch_size]

    for _, row in batch.iterrows():
        data = {
            "Timestamp": row["Timestamp"],
            "Temperature": row["Temperature"],
            "pH": row["pH"],
            "Dissolved Oxygen": row["Dissolved Oxygen"]
         }

        producer.send("water-stream", data)

    print(f"Sent batch {i} to {i+batch_size}")

producer.flush()
print("Data streaming completed")
```

```python
from kafka import KafkaConsumer
import json

consumer = KafkaConsumer(
    "water-stream",
    bootstrap_servers='localhost:9092',
    auto_offset_reset='earliest',
    enable_auto_commit=True,
    group_id="water-group",
    value_deserializer=lambda x: json.loads(x.decode('utf-8'))
)

count = 0

for msg in consumer:
    print(f"Partition: {msg.partition} | Offset: {msg.offset} | Value: {msg.value}")
    count += 1

    if count == 10:
        break
```
