```python
import pandas as pd
import os

path = r"C:\Water_Monitoring_System\streaming\real_streaming_input"
os.makedirs(path, exist_ok=True)

df = pd.read_csv(r"C:\Water_Monitoring_System\data\brisbane_water_quality.csv")

chunk_size = 500
rows = len(df)

for i in range(0, rows, chunk_size):
    file_path = f"{path}/data_{i}.csv"

    df.iloc[i:i+chunk_size].to_csv(file_path, index=False)

    print("Saved:", file_path)
```

```python
import os

files = sorted(os.listdir(path))
print(files)
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .master("local[*]") \
    .appName("StreamingSimulation") \
    .getOrCreate()
```

```python
import time
import os

for file in files:

    print(f"\n Processing: {file}")

    file_path = os.path.join(path, file)

    batch_df = spark.read.csv(
        file_path,
        header=True,
        inferSchema=True
    )
```
