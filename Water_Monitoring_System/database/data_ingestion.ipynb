```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine("postgresql://postgres:#Twinkle65@localhost:5432/water_db")
```

```python
def batch_ingestion(file_path, table_name, batch_size=1000):
    chunks = pd.read_csv(file_path, chunksize=batch_size)
    
    for i, chunk in enumerate(chunks):
        print(f"Ingesting batch {i+1}")
        
        # basic cleaning (optional)
        chunk.fillna(chunk.mean(numeric_only=True), inplace=True)
        chunk.drop_duplicates(inplace=True)
        
        # load into database
        chunk.to_sql(table_name, engine, if_exists='append', index=False)

    print("Batch ingestion completed")
```

```python
batch_ingestion(
    r"C:\Water_Monitoring_System\data\brisbane_water_quality.csv",
    "water_data_batch",
    batch_size=1000
)
```
