```python
from sqlalchemy import create_engine

engine = create_engine("postgresql://postgres:#Twinkle65@localhost:5432/water_db")
```

```python
import pandas as pd

df = pd.read_csv(r"C:\Water_Monitoring_System\data\brisbane_water_quality.csv")

df.head()
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
df.drop_duplicates(inplace=True)
```

```python
df.to_sql("water_data", engine, if_exists="replace", index=False)

print("Data uploaded successfully")
```

```python
pd.read_sql("SELECT * FROM water_data LIMIT 5", engine)
```

```python
# WHERE
pd.read_sql('SELECT * FROM water_data WHERE "Turbidity" > 2', engine)
```

```python
# GROUP BY
pd.read_sql('''
SELECT "Timestamp", AVG("Turbidity")
FROM water_data
GROUP BY "Timestamp"
''', engine)
```

```python
# ORDER BY
pd.read_sql('''
SELECT * FROM water_data
ORDER BY "Timestamp" DESC
''', engine)
```

```python
# Subquery
pd.read_sql('''
SELECT * FROM water_data
WHERE "Turbidity" > (
    SELECT AVG("Turbidity") FROM water_data
)
''', engine)
```

```python
# Creating Second table 
with engine.connect() as conn:
    conn.execute(text('''
    CREATE TABLE time_info (
        "Timestamp" TIMESTAMP,
        "Category" VARCHAR(50)
    )
    '''))
    conn.commit()
```

```python
with engine.connect() as conn:
    conn.execute(text(f'''
    INSERT INTO time_info ("Timestamp", "Category") VALUES
    ('{df["Timestamp"][0]}', 'Normal'),
    ('{df["Timestamp"][1]}', 'Alert'),
    ('{df["Timestamp"][2]}', 'Critical')
    '''))
    conn.commit()
```

```python
# Left Join
pd.read_sql('''
SELECT w."Timestamp", w."Turbidity", t."Category"
FROM water_data w
JOIN time_info t
ON w."Timestamp" = t."Timestamp"::text
''', engine)
```

```python
# Window functions 
pd.read_sql('''
SELECT 
    "Timestamp",
    "Turbidity",
    RANK() OVER (
        ORDER BY "Turbidity" DESC
    ) AS rank
FROM water_data
''', engine)
```

```python
 # OLAP
from sqlalchemy import text

with engine.connect() as conn:
    conn.execute(text('''
    CREATE TABLE water_summary AS
    SELECT 
        DATE("Timestamp"::timestamp) AS date,
        AVG("Turbidity") AS avg_turbidity,
        AVG("Temperature") AS avg_temperature,
        AVG("pH") AS avg_ph
    FROM water_data
    GROUP BY DATE("Timestamp"::timestamp)
    '''))
    conn.commit()
```

```python
pd.read_sql('SELECT * FROM water_summary', engine)
```
